```python
import matplotlib.pyplot
import numpy
import glob
```

```python
for file in glob.glob('inflammation-*.csv'):
    data = numpy.loadtxt(fname=file, delimiter=',')
    
    fig = matplotlib.pyplot.figure(figsize=(10, 3))
    matplotlib.pyplot.title(file)
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.set_ylabel('average')
    axes1.plot(numpy.mean(data, axis=0))

    axes2.set_ylabel('max')
    axes2.plot(numpy.max(data, axis=0))

    axes3.set_ylabel('min')
    axes3.plot(numpy.min(data, axis=0))

    fig.tight_layout()
    matplotlib.pyplot.show(fig)
```

```python
def load_data(pizza):
    """Loads a filename into a numpy array
    """
    data = np.loadtxt(fname=pizza, delimiter=',')
    return(data)
```

```python
load_data('inflammation-01.csv')
```

```python
def plot_data(data, plot_name):
    """Takes a numpy array, and a string for a plot
    and returns a matplotlib figure
    """
    fig = plt.figure(figsize=(10, 3))
    plt.title(plot_name)

    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))

    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))

    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))

    fig.tight_layout()
    plt.show(fig)
```

```python
data = load_data('inflammation-01.csv')
plot_data(data, 'inflammation-01.csv')
```
